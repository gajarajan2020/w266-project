### Read Smiliarity Dataframe Rows and retrieve the target label with highest ratio

```python
similarity_response = []

# Extract unique rows lenght 
rows = sim_df.drop_duplicates().shape[0]


for i in range(rows):
            # select rows with given index
            index_df = sim_df.loc[list(sim_df.index)[i]]
            if type(index_df.drop_duplicates()) == pd.core.series.Series:
                output = pd.DataFrame(index_df.drop_duplicates().transpose()).astype(float).idxmax()[0]
            else:
                output = pd.DataFrame(index_df.drop_duplicates().transpose()).astype(float).idxmax()[0]            
            
            if type(index_df.drop_duplicates()) == pd.core.series.Series:
                similarity = pd.DataFrame(index_df.drop_duplicates().transpose()).astype(float).max()[0]
            else:
                similarity = pd.DataFrame(index_df.drop_duplicates().transpose()).astype(float).max()[0]
            

            similarity_response.append(
            {
                "input" : sim_df.index[i] 
                # extract target column with highest similarity 
                , "output": output 
                # similiarity of between input and output
                , "similarity": similarity
            }
            )
```

```python
def parse(x):
    new_list=[]
    if list(x[0].values())[2] > 0.8:
        new_list.append(x)
    return new_list
parse(sorted(similarity_response, key=lambda x: list(x.values())[2], reverse=True))
```
